```python
def aggregate_method(code_lines):
    no_of_elements = 0
    ammortized_cost = 0
    operations = 0
    for i in code_lines:
        op = ""
        if "push" in i:
            op = "push"
            ammortized_cost+=1
            operations+=1
            no_of_elements+=1
        elif "multipop" in i:
            op = "multipop"
            times = int(i[i.index('(')+1])
            operations+=1
            for i in range(times):
                if no_of_elements>0:
                    no_of_elements-=1
                if no_of_elements>0:
                    ammortized_cost+=1
        elif "pop" in i:
            op = "pop"
            operations+=1
            if no_of_elements>0:
                no_of_elements-=1
            if no_of_elements>0:
                ammortized_cost+=1
        elif "peek" in i:
            op = "peek"
            ammortized_cost+=1
            operations+=1
            no_of_elements+=0
    return ammortized_cost,operations

def potential_method(code_lines):
    no_of_elements = 0
    ammortized_cost = 0
    operations = 0
    for i in code_lines:
        op = ""
        if "push" in i:
            op = "push"
            ammortized_cost+=2
            operations+=1
            no_of_elements+=1
        elif "multipop" in i:
            op = "multipop"
            times = int(i[i.index('(')+1])
            operations+=1
            for i in range(times):
                ammortized_cost+=0
                if no_of_elements>0:
                    no_of_elements-=1
        elif "pop" in i:
            op = "pop"
            ammortized_cost+=0
            operations+=1
            if no_of_elements>0:
                no_of_elements-=1
        elif "peek" in i:
            op = "peek"
            ammortized_cost+=1
            operations+=1
            no_of_elements+=0
    return ammortized_cost,operations

def accounting_method(code_lines):
#     Each operation is assigned with a cost of 2
    op_cost=2
    no_of_elements = 0
    ammortized_cost = 0
    operations = 0
    balance=0
    for i in code_lines:
        op = ""
        if "push" in i:
            op = "push"
            ammortized_cost+=1
            balance+=(op_cost-1)
            operations+=1
            no_of_elements+=1
        elif "multipop" in i:
            op = "multipop"
            times = int(i[i.index('(')+1])
            operations+=1
            for i in range(times):
                if no_of_elements>0:
                    no_of_elements-=1
                if no_of_elements>0:
                    ammortized_cost+=1
                balance-=1
            balance+=op_cost
        elif "pop" in i:
            op = "pop"
            balance+=(op_cost-1)
            operations+=1
            if no_of_elements>0:
                no_of_elements-=1
            if no_of_elements>0:
                ammortized_cost+=1
        elif "peek" in i:
            op = "peek"
            ammortized_cost+=1
            balance+=(op_cost-1)
            operations+=1
            no_of_elements+=0
    return ammortized_cost,operations,balance

def find_nth(haystack, needle, n):
    start = haystack.find(needle)
    while start >= 0 and n > 1:
        start = haystack.find(needle, start+len(needle))
        n -= 1
    return start
```
